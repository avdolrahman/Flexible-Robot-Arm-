```python
import matplotlib.pyplot as plt
import numpy as np

# Data
alpha = [1.81, 3.63, 5.44, -1.81, -3.63, -5.44]  # Alpha in degrees (y-axis)
tension = [1.5, 2.7, 3.8, -1, -2.2, -3.3]  # Tension in volts (x-axis)

# Fit a line to the data (Linear Regression)
slope, intercept = np.polyfit(tension, alpha, 1)  # Fit the line (1 = linear)

# Generate the fit line for plotting
fit_line = np.polyval([slope, intercept], tension)

# Create the plot
plt.figure(figsize=(8, 6))
plt.scatter(tension, alpha, color='blue', label='Data points')  # Scatter plot for the data
plt.plot(tension, fit_line, color='red', label=f'Fit line: $K_e$ = {slope:.2f}')  # Plot the fit line

# Add labels and title
plt.xlabel('Tension (V)', fontsize=12)
plt.ylabel(r'$\alpha$ (degree)', fontsize=12)
plt.title(r'Plot of $\alpha$ (degree) vs Tension (V)', fontsize=14)

# Add a grid and a legend
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
